```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
df = pd.read_csv("41750-15-episode_total_reward.csv")
df2 = pd.read_csv("70159-25-episode_total_reward.csv")
df25 = pd.read_csv("49436-35-episode_total_reward.csv")
df35 = pd.read_csv("44485-45-episode_total_reward.csv")
df45 = pd.read_csv("47668-55-episode_total_reward.csv")


plt.figure(figsize=(20, 10))
y_lowess = sm.nonparametric.lowess(df["Value"], df["Step"],frac  = 0.10)  # 10 % lowess smoothing
y_lowess2 = sm.nonparametric.lowess(df2["Value"], df2["Step"],frac  = 0.10)  # 10 % lowess smoothing
y_lowess25 = sm.nonparametric.lowess(df25["Value"], df25["Step"],frac  = 0.10)  # 10 % lowess smoothing
y_lowess35 = sm.nonparametric.lowess(df35["Value"], df35["Step"],frac  = 0.10)  # 10 % lowess smoothing
y_lowess45 = sm.nonparametric.lowess(df45["Value"], df45["Step"],frac  = 0.10)  # 10 % lowess smoothing

plt.plot(y_lowess25[:, 0], y_lowess25[:, 1], color = "green")
plt.plot(y_lowess35[:, 0], y_lowess35[:, 1], color = "purple")
plt.plot(y_lowess45[:, 0], y_lowess45[:, 1], color = "orange")
plt.plot(y_lowess[:, 0], y_lowess[:, 1], color = "blue")
plt.plot(y_lowess2[:, 0], y_lowess2[:, 1], color = "red")

plt.legend(["1/5 Reinforcement Learning agent", "2/5 Reinforcement Learning agent", "3/5 Reinforcement Learning agent", "4/5 Reinforcement Learning agent", "5/5 Reinforcement Learning agent"], loc='lower right', fontsize ='xx-large')


plt.xlabel("Episode")
plt.ylabel("Reward")
plt.savefig("img.jpg")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm

df1 = pd.read_csv("DQN-2000-episode_total_reward.csv")
df2 = pd.read_csv("DDQN-2000-episode_total_reward.csv")


plt.figure(figsize=(20, 10))
y_lowess = sm.nonparametric.lowess(df1["Value"], df1["Step"],frac  = 0.10)  # 10 % lowess smoothing
y_lowess2 = sm.nonparametric.lowess(df2["Value"], df2["Step"],frac  = 0.10)  # 10 % lowess smoothing

plt.plot([12.3 for _ in range(2500)], linestyle="--")

plt.plot(y_lowess[:, 0], y_lowess[:, 1], color = "blue")
plt.plot(y_lowess2[:, 0], y_lowess2[:, 1], color = "green")

plt.legend(["Deep Q-Network - 5/5 CAV", "Double Deep Q-NetWork - 5/5 CAV"], loc='lower right', fontsize ='xx-large')


plt.xlabel("Episode")
plt.ylabel("Sum reward")
plt.savefig("55_2000.jpg")
```
